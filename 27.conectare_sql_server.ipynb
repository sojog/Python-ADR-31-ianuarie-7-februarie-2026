```python
# /Users/mb_pro/Documents/Bittnet/15.Python-ianuarie-31-ian-7-feb/Cod/Python-ADR-31-ianuarie-7-februarie-2026/sql_server_list_databases.py

import pyodbc

# Informații de conectare
server = 'NumeleTăuDeServer'
database = 'master'  # De obicei ne conectăm la DB master pentru a lista toate bazele de date
username = 'UtilizatorulTau'
password = 'ParolaTa'

# Conectarea la server
try:
    connection_string = f'DRIVER={{ODBC Driver 17 for SQL Server}};SERVER={server};DATABASE={database};UID={username};PWD={password}'
    conn = pyodbc.connect(connection_string)
    print("Conexiune reușită!")

    # Crearea unui cursor
    cursor = conn.cursor()

    # Executarea unei cereri pentru a lista toate bazele de date
    cursor.execute('SELECT name FROM sys.databases')
    databases = cursor.fetchall()

    print("Bazele de date disponibile:")
    for row in databases:
        print(row.name)

except Exception as e:
    print(f"Eroare: {e}")

finally:
    if conn:
        conn.close()
        print("Conexiune închisă.")
```
