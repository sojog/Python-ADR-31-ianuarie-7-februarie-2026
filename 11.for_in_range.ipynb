In alte limbaje: for (int i = 0; i < 10; i ++)

```python
for i in range(10):
    print("i=", i)
```

```python
for i in range(5, 10):
    print("i=", i)
```

```python
for i in range(50, 60):
    print("i=", i, type(i))
```

```python
list(range(50, 60))
```

```python
for i in range(50, 60, 2):
    print("i=", i, type(i))
```

```python
for i in range(50, 60, 3):
    print("i=", i, type(i))
```

```python
for i in range(100, 90):
    print("i=", i, type(i))
```

```python
for i in range(100, 90, -1):
    print("i=", i, type(i))
```
