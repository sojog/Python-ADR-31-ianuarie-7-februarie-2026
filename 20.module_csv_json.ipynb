```python
import json
```

```python
dictionar = {
    "cer" : "albastru",
    "iarba" : "verde",
    "soare" : "galben"
}

with open("fisier_json.json", "w") as file_writer:
    json.dump(dictionar, file_writer )
```

```python
with open("fisier_json.json", "r") as file_reader:
    obiect_citit = json.load(file_reader)
    print(obiect_citit, type(obiect_citit))
```

```python
import csv
```

```python
dictionar = {
    "cer" : "albastru",
    "iarba" : "verde",
    "soare" : "galben"
}

with open("fisier_csv.csv", "w") as file_writer:
    csv_writer = csv.DictWriter(file_writer, dictionar.keys())
    csv_writer.writerow(dictionar)
```

```python

with open("fisier_csv.csv", "r") as file_reader:
    csv_reader = csv.DictReader(file_reader, dictionar.keys())
    for row in csv_reader:
        print("row:", row, type(row))
```
