```python
# Lista este o secventa (colectie) heterogena
lista = [20, 'hello', 30, 40, 32.032, True, 'hello', [32, 321, 312], 20]

print(type(lista))
for i in lista:
    print("i=", i)
```

```python
# Tuple este o secventa (colectie) heterogena
tuplu = (20, 'hello', 30, 40, 32.032, True, 'hello', [32, 321, 312], 20)

for i in tuplu:
    print("i=", i)
```
