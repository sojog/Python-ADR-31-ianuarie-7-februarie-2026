```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome()
```

```python
browser.get("https://www.cursbnr.ro/")
```

```python
from selenium.webdriver.common.by import By
```

```python
AGREE_XPATH = "/html/body/div[6]/div[2]/div[2]/div[2]/div[2]/button[1]/p"
```

```python
agree_button = browser.find_element(By.XPATH, AGREE_XPATH)
```

```python
agree_button.click()
```

```python
EURO_XPATH = "/html/body/div[3]/div[1]/div/main/div[2]/div[1]/ul/li[1]/div[1]/div[1]"
```

```python
euro_label = browser.find_element(By.XPATH, EURO_XPATH)
euro_label
```

```python
euro_label.text
```

```python
# w -write (rescriere daca exista)
# r -read
# a-append (adauga la final)
with open("euro.txt", "w") as file_handler:
    file_handler.write(euro_label.text)

```

```python
SUMA_EXPATH = '//*[@id="initial-value"]'
```

```python
suma_input = browser.find_element(By.XPATH, SUMA_EXPATH)
suma_input
```
